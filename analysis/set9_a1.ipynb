```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path_merge_random = 'MergeSort_random.csv'
data_merge_random = pd.read_csv(file_path_merge_random, delimiter=',')

size_merge_random = data_merge_random['Size']
time_merge_random = data_merge_random['Time(ms)']
comparisons_merge_random = data_merge_random['Comparisons']

file_path_merge_reversed = 'MergeSort_reversed.csv'
data_merge_reversed = pd.read_csv(file_path_merge_reversed, delimiter=',')

size_merge_reversed = data_merge_reversed['Size']
time_merge_reversed = data_merge_reversed['Time(ms)']
comparisons_merge_reversed = data_merge_reversed['Comparisons']

file_path_merge_almostSorted = 'MergeSort_almostSorted.csv'
data_merge_almostSorted = pd.read_csv(file_path_merge_almostSorted, delimiter=',')

size_merge_almostSorted = data_merge_almostSorted['Size']
time_merge_almostSorted = data_merge_almostSorted['Time(ms)']
comparisons_merge_almostSorted = data_merge_almostSorted['Comparisons']
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(size_merge_random, time_merge_random, color='black', linestyle='-', marker='o')
plt.scatter(size_quick_random, time_quick_random, color='red', linestyle='-', marker='o')

plt.plot(size_quick_random, time_quick_random, color='red', linestyle='-', label='Quick Sort')
plt.plot(size_merge_random, time_merge_random, color='black', linestyle='-', label='Merge Sort')

plt.title('Merge sort VS Quick sort, random arrays')
plt.xlabel('Array size')
plt.ylabel('Time, ms')
plt.grid(True)
plt.legend()

plt.show()
```
